```python
##Entrada de cantidad de conjuntos
conjunt= int(input("Ingrese la cantidad de conjuntos que desea crear: "))
##Entrada de  cantidad de elementos
while(conjunt<2):#ciclo para invalidar opcion y escoger opcion dentro del rango
  print("INVALIDO. Debe tener un minimo de 2 conjuntos.")
  conjunt= int(input("Ingrese la cantidad de conjuntos que desea crear: "))
conjuntu= 0 #Inicializa contador de los conjuntos
#Declaracion del arreglo que almacena conjunto
C={} #arreglo que almacena los conjuntos del usuario

while(conjunt>=2 and conjuntu<conjunt): #bucle para almacenar los conjuntos
  conjuntu= conjuntu+1 #indica el numero de conjunto en formación
  elem= int(input("Ingrese la cantidad de elementos que desea tener: "))
  print("Como desea crear este conjunto?") #Eleccion de forma de creacion
  deci=int(input("Menu: \n1. Por llaves\n2. Por string\n3. Por lista\n"))

  #ciclo para invalidar opcion y escoger opcion dentro del rango
  while(deci != 1 and deci != 2 and deci != 3):
    print("INVALIDO. Ingrese una opcion dentro del rango.")
    deci=int(input("Menu: \n1. Por llaves\n2. Por string\n3. Por lista\n"))

  conjunto= {0} #declaracion del conjunto
  #se limpia el conjunto para que se puedan ingresar nuevos elementos
  conjunto.clear()
  print("ADVERTENCIA. Si usted repite datos con el mismo valor para los")
  print("conjuntos, tenga en cuenta que solo se tomará en cuenta uno")
  if(deci==1):#crea el conjunto por llave
    print("Usted ha seleccionado trabajar por llaves")
    tamf= elem
    #ciclo que recoge los elementos del conjunto hasta que no quede 
    #espacio en el conjunto
    while(tamf>0):
      print("Hay espacio para ", tamf, "elementos.")
      n=input("Ingrese el elemento: ")
      conjunto.add(n)#se encarga de almacenar los datos nuevos en el conjunto
      tamcon= int(len(conjunto)) #Cuenta los elementos totales en el conjunto
      tamf= elem-tamcon #calcula la cantidad de espacio libre en el conjunto
  elif(deci==2):#crea el conjunto por string
    print("Usted ha seleccionado trabajar por string")
    n=set(input("Ingrese la palabra o frase string: "))
    tamcon=len(n) #cuenta los elementos totales en el conjunto

    #calcula si la cantidad de elementos ingresados es mayor al espacio 
    #solicitado
    tamf= elem-tamcon 

    while(tamf<0):#condicion para ingresar una frase o palabra mas corta
      print("Esta palabra o frase es mayor a la cantidad de elementos.")
      #entrada de una frase o palabra mas corta
      n= set(input("Ingrese otra palabra o frase string: "))
      #calcular el tamaño de la frase/palabra que el usuario desea ingresar
      tamcon=len(n)
      #calcular si la cantidad de elementos ingresados es mayor al solicitado
      tamf= elem-tamcon

    #condicion para almacenar palabra o frase y calcular los espacios que 
    #sobran
    while(tamf<elem and tamf>0):
      #almacenar frase/palabra ingresada
      conjunto.update(n)
      #imprimir cantidad de espacios sobrantes
      print("Faltan ", tamf,"elementos por ingresar.")
      #entradas de elementos que completen el conjunto
      n= set(input("Ingrese una palabra o frase string: "))
      #cacula si la cantidad de elementos nuevos sobrepasa la cantidad de
      #espacio sobrante
      tamn= len(conjunto) + len(n)
      
      #condicion para cambiar la frase ingresada en caso de que 
      #sea muy larga para el espacio que queda
      while(tamn>elem): 
        #Aviso de que la palabra/frase es muy larga para completar
        print("La ultima frase ingresada es muy larga.")
        #Entrada de una palabra/frase mas corta
        n= set(input("Ingrese una palabra o frase string: "))
        #cacula si la cantidad de elementos nuevos sobrepasa la cantidad de
        #espacio sobrante
        tamn= len(conjunto) + len(n)
      #condicion para almacenar palabra o frase y calcular los 
      #espacios que sobran
      if(tamn<=elem):
       #almacenar palabra/frase porque no sobrepasa la cantidad de 
       #elementos solicitados
       conjunto.update(n)
       #calcular cantidad de elementos del conjunto
       tamcon=len(conjunto)
       #calcular si quedan espacios en el conjunto o no.
       tamf= elem-tamcon
    #condicion para almacenar palabra o frase en caso de que sea exacta
    # al espacio brindado
    if(tamf==0):
      conjunto.update(n)

  elif(deci==3):#crea el conjunto por listas
    print("Usted ha seleccionado trabajar por listas")
    #guardar cantidad de elementos solicitados en otra variable
    tamf= elem
    #ciclo para almacenar datos hasta que se acaben los espacios sobrantes
    while(tamf>0):
      #Imprime la cantidad de espacio sobrante en el conjunto
      print("Hay espacio para ", tamf, "elementos.") 
      n=set(input(["Ingrese el elemento: "]))
      #se encarga de almacenar los nuevos elementos ingresados al conjunto
      conjunto.update(n)
      #cuenta la cantidad de elementos en el conjunto
      tamcon= int(len(conjunto))
      tamf= elem-tamcon#calcula la cantidad de espacio sobrante para elementos
      
  C[conjuntu]= conjunto
  #Imprime el numero de conjunto que se ha creado
  print("Conjunto no.", conjuntu) 
  #Imprime los elementos del conjunto
  print("El conjunto ingresado esta conformado por: ",conjunto)
  #Imprime la cantidad de elementos del conjunto 
  print("Este conjunto terminó con: ",len(conjunto) ,"elementos") 

#ingreso de decision del usuario para realizar operacioes
op=int(input("Desea evaluar una operacion?\n1. Si\n2. No\n")) 
#ciclo que evalua las operaciones si el usuario acepta hacer las operaciones

while(op!= 1 and op!= 2):
  print("INVALIDO. Ingrese una opcion dentro del rango(1-2)")
  op=int(input("Desea evaluar una operacion?\n1. Si\n2. No\n")) 

resultados={} #arreglo encargado de almacenar los resultados de las operaciones
r = 0 #indice de resultados
while(op==1): #bucle para realizar operaciones hasta que el usuario desee parar
  
  print("ADVERTENCIA: El resultado de cada operacion sera considerado como un")
  print("nuevo conjunto y será agregado a la lista de conjuntos con los que")
  print("puede trabajar.\n")
  print("¿Que operacion desea realizar. Escoja una opción del menú:")
  print("1. Unión de conjuntos \n2. Intersección de conjuntos")
  print("3. Diferencia de conjuntos\n4. Diferencia simétrica de conjuntos")
  #entrada de la opcion que desea el usuario
  ope=int(input("5. Comprobar conjunto Mayor\n6. Comprobar conjunto Menor\n"))

  #ciclo para invalidar opcion y escoger opcion dentro del rango
  if(ope!=1 and ope!=2 and ope!=3 and ope!=4 and ope!=5 and ope!=6):
    print("OPCION NO VALIDA. INGRESE UN NUMERO DENTRO DEL RANGO.")
  else:#condicion para que pueda recorrer el resto del programa si cumple
    print("Sus conjuntos estan organizados de la siguiente forma: ")
    for conjuntu in C: #Impresion de los conjuntos actuales
      print(conjuntu,")=", C[conjuntu])
    #Entrada de los conjuntos escogidos
    #Almacena conjunto principal
    A= int(input("Escoja el primer conjunto a evaluar: ")) 
    #Almacena el segundo conjunto a utilizar
    B= int(input("Escoja el segundo conjunto a evaluar: "))

    #bucle que invalida opciones inexistentes y da la oportunidad de ingresar
    #nuevas opciones que esten dentr del rango
    while((A<0 or A>conjuntu)or(B<0 or B>conjuntu)):
      print("Una de sus opciones es inválida.")
      print("Escoja un número dentro del rango.")
      A= int(input("Escoja el primer conjunto a evaluar: "))
      B= int(input("Escoja el segundo conjunto a evaluar: "))

    if(ope==1): #Operación de conjuntos – Unión.
      print("Usted ha escogido: Unión de conjuntos")
      NC= C[A].union(C[B]) #operacion que une conjuntos
    elif(ope==2): #Operación de conjuntos – Intersección.
      print("Usted ha escogido: Intersección de conjuntos.")
      NC= C[A] & C[B] #operacion que intersecta conjuntos
    elif(ope==3): #Operación de conjuntos – Diferencia.
      print("Usted ha escogido: Diferencia de conjuntos.")
      #operacion encargada de calcular la diferencia 
      #entre conjuntos
      NC= C[A].difference(C[B]) 
    elif(ope==4): #Operación de conjuntos – Diferencia Simétrica.
      print("Usted ha escogido: Diferencia simétrica de conjuntos.")
      #operacion encargada de calcular la diferencia simétrica
      #entre conjuntos
      NC= C[A].symmetric_difference(C[B])
    elif(ope==5): #Operación de conjuntos – Comprobar Conjunto Mayor.
      print("Usted ha escogido: Comprobar conjunto Mayor")
      #operación encargada de evaluar si el conjunto principal es mayor o no
      NC= C[A].issuperset(C[B])
    elif(ope==6): #Operación de conjuntos – Comprobar Conjunto Menor.
      print("Usted ha escogido: Comprobar conjunto Menor")
      #operación encargada de evaluar si el conjunto principal es menor o no
      NC= C[A].issubset(C[B])
    
    #imprime el resultado de la operacion realizada
    print("El resultado de esta operacion es: ", NC)

    #indica la posicion del resultado dentro del arreglo que los almacena
    r = r + 1
    #asigna el resultado a la posicion que le toca dentro del arreglo
    resultados[r]= NC

    #condición para que los resultados que no sean conjuntos no sean contados
    #como posibles conjuntos para evaluar operaciones
    if(ope>=1 and ope<=4 and NC!= set()):
      conjuntu= conjuntu+1
      C[conjuntu]= NC
    
    #Le permite al usuario seguir realizando operaciones o terminar
    op=int(input("Desea evaluar otra operacion?\n1. Si\n2. No\n"))
    #ciclo para invalidar opciones y obliga al usuario a escoger una opción 
    #dentro del rango
    while(op!= 1 and op!=2):
      print("INVALIDO. INGRESE UNA OPCION DENTRO DEL RANGO")
      op=int(input("Desea evaluar otra operacion?\n1. Si\n2. No\n"))


#Parte final de la impresion
print("Se han ingresado todos los conjuntos.")
print("Sus conjuntos originales son: ")

for conjuntu in C: #bucle encargado de imprimir los conjuntos originales 
  #bucle que condiciona al bucle para que no imprima los conjuntos provenientes
  #de operaciones
  if(conjuntu<=conjunt): 
    print(conjuntu,")=", C[conjuntu])
if(r>0):
  print("Los resultados obtenidos de sus operaciones son: ")
  #bucle encargado de imprimir todas las respuestas obtenidas, sin importar
  #si son conjuntos o no
  for r in resultados:
    print(r,")=", resultados[r])

print("Se ha terminado el Proyecto no.1") #Finalizacion del proyecto no.1
```
